```python
#MISSING VALUE CORRECTION
```

```python
# Fill missing values using station-wise mean for each pollutant column
cols_to_fill = ['pollutant_min', 'pollutant_max', 'pollutant_avg']

for col in cols_to_fill:
    df[col] = df.groupby('station')[col].transform(lambda x: x.fillna(x.mean()))
```

```python
df[['pollutant_min', 'pollutant_max', 'pollutant_avg']].isna().sum()
```

```python
df['pollutant_min'].fillna(df['pollutant_min'].mean(), inplace=True)
df['pollutant_max'].fillna(df['pollutant_max'].mean(), inplace=True)
df['pollutant_avg'].fillna(df['pollutant_avg'].mean(), inplace=True)
```

```python
df[['pollutant_min', 'pollutant_max', 'pollutant_avg']].isna().sum()
```

```python
#CREATING DERIVED COLUMNS
```

```python
def pollutant_type(pid):
    if pid in ['PM2.5', 'PM10']:
        return 'Particulate'
    else:
        return 'Gas'

df['pollutant_type'] = df['pollutant_id'].apply(pollutant_type)
```

```python
state_mean_pollution = df.groupby('state')['pollutant_avg'].mean().sort_values(ascending=False)
state_mean_pollution
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14,7))
ax = sns.barplot(
    x=state_mean_pollution.index,
    y=state_mean_pollution.values,
    palette='coolwarm'
)

plt.title("Mean Pollution Score per State")
plt.xlabel("State")
plt.ylabel("Mean Pollutant Average")
plt.xticks(rotation=90)

# Add value labels on top of each bar
for i, v in enumerate(state_mean_pollution.values):
    ax.text(i, v + 0.5, f"{v:.2f}", ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# Step 1: Calculate mean pollution per city
city_mean = df.groupby('city')['pollutant_avg'].mean().reset_index()

# Step 2: Merge with original df to get state info
city_state = df[['city', 'state']].drop_duplicates()

merged = city_mean.merge(city_state, on='city', how='left')

# Step 3: Sort and take top 10
top_10_city_state = merged.sort_values(by='pollutant_avg', ascending=False).head(10)

top_10_city_state
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
ax = sns.barplot(
    data=top_10_city_state,
    x='city',
    y='pollutant_avg',
    hue='state',
    palette='tab10'
)

plt.title("Top 10 Most Polluted Cities with Their States")
plt.xlabel("City")
plt.ylabel("Mean Pollutant Average")
plt.xticks(rotation=45)

# Add value labels
for i, v in enumerate(top_10_city_state['pollutant_avg']):
    ax.text(i, v + 0.5, f"{v:.2f}", ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# Step 1: Calculate mean pollution per city
city_mean = df.groupby('city')['pollutant_avg'].mean().reset_index()

# Step 2: Merge with state information
city_state = df[['city', 'state']].drop_duplicates()

merged = city_mean.merge(city_state, on='city', how='left')

# Step 3: Sort ascending (lowest pollution first)
bottom_10_city_state = merged.sort_values(by='pollutant_avg', ascending=True).head(10)

bottom_10_city_state
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
ax = sns.barplot(
    data=bottom_10_city_state,
    x='city',
    y='pollutant_avg',
    hue='state',
    palette='Greens'
)

plt.title("Bottom 10 Least Polluted Cities with Their States")
plt.xlabel("City")
plt.ylabel("Mean Pollutant Average")
plt.xticks(rotation=45)

# Add value labels
for i, v in enumerate(bottom_10_city_state['pollutant_avg']):
    ax.text(i, v + 0.3, f"{v:.2f}", ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```
